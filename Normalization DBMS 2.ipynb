```python
#3NF
import pandas as pd
data = {
    'Student_ID' : [1,2,3,4],
    'Student_Name' : ['Alice','Bob','Charlie','David'],
    'Course' : ['Math','Physics','Math','Chemistry'],
    'Professor':['Dr.Smith','Dr.Jhonson','Dr.Smith','Dr.Davis'],
    'Credit_Hours' : [3,4,3,5],
    'Room_Number' : [101,203,101,305]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

#create student dataFrame
student_df = df[['Student_ID','Student_Name']].drop_duplicates().reset_index(drop=True)
print("\n")
print(student_df)

courses_df = df[['Course','Credit_Hours']].drop_duplicates().reset_index(drop=True)
print("\n")
print(courses_df)

professor_df = df[['Professor','Room_Number']].drop_duplicates().reset_index(drop=True)
print("\n")
print(professor_df)

student_course_df = df[['Student_ID','Course']].drop_duplicates().reset_index(drop=True)
print("\n")
print(student_course_df)
```

```python
#3NF
import pandas as pd

# Original DataFrame
data = {
    'Student_Id' : ['1DT15ENG01','1DT15ENG02','1DT15ENG03','1DT15ENG04'],
    'Student_Name' : ['Alex','Barry','Clair','David'],
    'Subject_Id' : ['15CS11','15CS13','15CS12','15CS13'],
    'Subject' : ['SQL','JAVA','C++','JAVA'],
    'Address' : ['Goa','Bengaluru','Delhi','Kochi']
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Splitting into Student_Info
student_info = df[['Student_Id', 'Student_Name', 'Address']].drop_duplicates().reset_index(drop=True)

# Splitting into Subject_Info
subject_info = df[['Subject_Id', 'Subject']].drop_duplicates().reset_index(drop=True)

# Creating Student_Subject relation table
student_subject = df[['Student_Id', 'Subject_Id']]

# Print the resulting DataFrames
print("\nStudent_Info:")
print(student_info)
print("\nSubject_Info:")
print(subject_info)
print("\nStudent_Subject:")
print(student_subject)
```

```python
#BCNF
import pandas as pd

data = {
    'Emp_Id': [1, 2, 3, 4, 5],
    'Emp_Name': ['Alex', 'Bob', 'Charlie', 'David', 'Eva'],
    'Proj_Id': [101, 102, 101, 103, 102],
    'Proj_Name': ['ProjectA', 'ProjectB', 'ProjectA', 'ProjectC', 'ProjectB'],
    'Dept_Id': [201, 202, 201, 203, 202],
    'Dept_Name': ['HR', 'IT', 'HR', 'Finance', 'IT'],
    'Manager_Id': [3, 2, 3, 4, 2],
    'Manager_Name': ['Charlie', 'Bob', 'Charlie', 'David', 'Bob']
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Emp_Info
emp_info = df[['Emp_Id', 'Emp_Name', 'Proj_Id', 'Dept_Id', 'Manager_Id']].drop_duplicates().reset_index(drop=True)

# Proj_Info
proj_info = df[['Proj_Id', 'Proj_Name']].drop_duplicates().reset_index(drop=True)

# Dept_Info
dept_info = df[['Dept_Id', 'Dept_Name']].drop_duplicates().reset_index(drop=True)

# Manager_Info
manager_info = df[['Manager_Id', 'Manager_Name']].drop_duplicates().reset_index(drop=True)

print("\nEmp_Info:")
print(emp_info)
print("\nProj_Info:")
print(proj_info)
print("\nDept_Info:")
print(dept_info)
print("\nManager_Info:")
print(manager_info)
```

```python
#4NF
import pandas as pd
data = {
    'Component_ID' : [1,2,3,4,5],
    'Component_Name' : ['Resistor','Transistor','Capacitor','IC Chip',''],
    'Supplier_ID' : [101,102,103,104,105],
    'Supplier_Name' : ['Electro_Supply','TransTech','capCorp','IC solution','Ak Memory'],
    'Plant_ID' : [201,202,203,204,205],
    'Plant_Location' : ['FactoryA','FactoryB','FactoryC','FactoryD','FactoryE'],
    'Production_ID' : [301,302,303,304,305],
    'Production_Date' : ['2022-01-10','2022-02-05','2022-03-20','2022-01-15','2022-09-06'],
    'Quality_control_ID' : [401,402,403,404,405],
    'Quality_CIntrol_Result' : ['pass','fail','pass','pass','fail']
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Component DF
component_df = df[['Component_ID', 'Component_Name']].drop_duplicates().reset_index(drop=True)

# Production run dataframe
production_df = df[['Production_ID', 'Production_Date']].drop_duplicates().reset_index(drop=True)

# Supplier dataFrame
supplier_df = df[['Supplier_ID', 'Supplier_Name']].drop_duplicates().reset_index(drop=True)


# Quality control dataFrame
quality_control_df = df[['Quality_control_ID', 'Quality_CIntrol_Result']].drop_duplicates().reset_index(drop=True)

# plant dataFrame
plant_df = df[['Plant_ID', 'Plant_Location']].drop_duplicates().reset_index(drop=True)

# component dataFrame
component_supplier_df = df[['Component_ID','Supplier_ID','Plant_ID', 'Production_ID','Quality_control_ID']].drop_duplicates().reset_index(drop=True)

print("\component_df:")
print(component_df)
print("\nProduction run dataframe:")
print(production_df)
print("\nSupplier dataFrame:")
print(supplier_df)
print("\nQuality control dataFrame")
print(quality_control_df)
print("\nplant dataFrame")
print(plant_df)
print("\ncomponent dataFrame")
print(component_supplier_df)
```
