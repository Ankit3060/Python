```python
#BCNF
import pandas as pd
data = {
    'E_ID' : [101,101,102,102],
    'City' : ['Delhi','Delhi','Mumbai','Mumbai'],
    'D_Name' : ['Accounting','Development','Admin','Testing'],
    'D_Type' : ['D10','D10','D20','D20'],
    'D_No' : [201,301,202,304]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)


# Emp_Info
emp_info = df[['E_ID', 'City', 'D_Name', 'D_Type']].drop_duplicates().reset_index(drop=True)

# Dept_Info
dept_info = df[['D_No', 'D_Name','D_Type']].drop_duplicates().reset_index(drop=True)

print("\nEmp_Info:")
print(emp_info)

print("\nDept_Info:")
print(dept_info)
```

```python
#BCNF
import pandas as pd
data = {
    'Professor_Code' : ['P1','P1','P2','P2','P3'],
    'Department' : ['Physics','Mathematics','Chemistry','Physics','Mathematics'],
    'HOD' : ['Gosh','Krishna','Rao','Gosh','Krishnan'],
    'percent_time' : [50,50,25,75,100]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Create a new relation based on the FD Professor_Code -> Department, HOD, percent_time
professor_info = df[['Professor_Code', 'Department', 'HOD', 'percent_time']].drop_duplicates().reset_index(drop=True)

# Create a relation for Department and HOD
department_hod_info = df[['Department', 'HOD']].drop_duplicates().reset_index(drop=True)

print("\nProfessor_Info:")
print(professor_info)

print("\nDepartment_HOD_Info:")
print(department_hod_info)
```

```python
#3NF
import pandas as pd
data = {
    'Roll_no' : [1784,1648,1768,1848,1682,1485],
    'Name' : ['Raman','Krishna','Gopalan','Raja','Maya','Singh'],
    'Department' : ['Physics','Chemistry','Mathematics','Botany','Geology','Zoology'],
    'Year' : [1,1,2,2,3,4],
    'Hostel_Name' : ['Ganga','Ganaga','Kaveri','Kaveri','Krishna','Godavari']
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

#1 Dependency diagram   [ Roll_no -> Name, Department, Year, Hostel_Name ]

#2 Show relation is in 2NF
# Primary key: Roll_no
# Non-prime attributes: Name, Department, Year, Hostel_Name

#3 3NF
# Relation 1: Roll_no, Name, Department
student_info = df[['Roll_no', 'Name', 'Department']].drop_duplicates().reset_index(drop=True)

# Relation 2: Year, Hostel_Name
hostel_info = df[['Year', 'Hostel_Name']].drop_duplicates().reset_index(drop=True)

print("\nStudent_Info:")
print(student_info)

print("\nHostel_Info:")
print(hostel_info)
```

```python
#BCNF
import pandas as pd

data = {
    'Insurance_No' : [1135,1057,1068,1140],
    'Contract_No' : ['C1024','C1024','C1025','C1025'],
    'HRS' : [16,20,24,16],
    'E_Name' : ['Sharma','Sahni','Gupta','Jindal'],
    'Hotel_No' : ['H25','H25','H04','H04'],
    'Location' : ['Delhi','Delhi','Mumbai','Mumbai']
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Relation 1: Insurance_No, Contract_No, HRS, E_Name
relation1 = df[['Insurance_No', 'Contract_No', 'HRS', 'E_Name']].drop_duplicates().reset_index(drop=True)

# Relation 2: Insurance_No, Contract_No, Hotel_No, Location
relation2 = df[['Insurance_No', 'Contract_No', 'Hotel_No', 'Location']].drop_duplicates().reset_index(drop=True)

print("\nRelation 1:")
print(relation1)

print("\nRelation 2:")
print(relation2)
```
